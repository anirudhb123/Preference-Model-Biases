```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import os

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    "font.size":       18,
    "axes.titlesize":  20,
    "axes.labelsize":  18,
    "xtick.labelsize": 18,
    "ytick.labelsize": 18,
    "legend.fontsize": 15
})

# enhanced colormap setup
a = plt.get_cmap("Dark2").colors
b = plt.get_cmap("Set2").colors
deleted = np.delete(a, 1, axis=0)
new_colors = np.insert(deleted, 1, b[1], axis=0)
cmap = ListedColormap(new_colors, name="Custom2")

# === CONFIG ===
DATA_ROOT = '../../../data'
HUMAN_ROOT = '../../../human_annotation_data'
BASE_DIR = os.path.join(DATA_ROOT, 'fine_tuned_model_scores')
FT_DIR   = os.path.join(DATA_ROOT, 'fine_tuned_model_scores', 'vagueness_jargon_length')
BIAS_LIST = ['vagueness', 'length', 'jargon']
SIZE_LIST = ['2B', '3B', '8B']
FT_SUFFIX = '500'

# helpers
def load_human_annotations(feat):
    df = pd.DataFrame(json.load(open(os.path.join(HUMAN_ROOT, feat, f'{feat}_grouped_human_annotations.json'), encoding='utf-8')))
    df = df[~df['most_frequent_preference'].str.lower().eq('tie')]
    col_map = {
        'vagueness': 'more_vague_response',
        'length':    'longer_response',
        'jargon':    'more_jargony_response'
    }
    df['human_pref'] = np.where(df['most_frequent_preference'] == df[col_map[feat]], 'perturbed', 'base')
    return df[['query','human_pref']]

def load_scores(path):
    return pd.DataFrame([json.loads(line) for line in open(path, encoding='utf-8') if line.strip()])

# compute metrics
human_skew = {}
base_means, base_stds = {}, {}
ft_means, ft_stds = {}, {}

for feat in BIAS_LIST:
    human = load_human_annotations(feat)
    human_skew[feat] = (human['human_pref']=='perturbed').mean()*100
    base_records, ft_records = [], []
    for size in SIZE_LIST:
        # base
        db = load_scores(os.path.join(BASE_DIR, feat, f'{feat}_scored_{size}.jsonl')).merge(human, on='query')
        pb = np.where(db['perturbed_score']>db['base_score'],'perturbed','base')
        base_records.append({
            'skew': (pb=='perturbed').mean()*100,
            'miscal': (pb!=db['human_pref']).mean()*100
        })
        # fine-tuned
        df_ft = load_scores(os.path.join(FT_DIR, f'{feat}_scored_{size}_{FT_SUFFIX}.jsonl')).merge(human, on='query')
        pf = np.where(df_ft['perturbed_score']>df_ft['base_score'],'perturbed','base')
        ft_records.append({
            'skew': (pf=='perturbed').mean()*100,
            'miscal': (pf!=df_ft['human_pref']).mean()*100
        })
    base_df = pd.DataFrame(base_records)
    ft_df   = pd.DataFrame(ft_records)
    base_means[feat] = base_df['skew'].mean()
    base_stds[feat]  = base_df['skew'].std()
    ft_means[feat]   = ft_df['skew'].mean()
    ft_stds[feat]    = ft_df['skew'].std()
    # store miscal too
    base_means[f'{feat}_miscal'] = base_df['miscal'].mean()
    base_stds[f'{feat}_miscal']  = base_df['miscal'].std()
    ft_means[f'{feat}_miscal']   = ft_df['miscal'].mean()
    ft_stds[f'{feat}_miscal']    = ft_df['miscal'].std()

# build DataFrames
bm_skew = pd.Series({f: base_means[f] for f in BIAS_LIST})
bs_skew = pd.Series({f: base_stds[f] for f in BIAS_LIST})
fm_skew = pd.Series({f: ft_means[f] for f in BIAS_LIST})
fs_skew = pd.Series({f: ft_stds[f] for f in BIAS_LIST})
bm_mis  = pd.Series({f: base_means[f+'_miscal'] for f in BIAS_LIST})
bs_mis  = pd.Series({f: base_stds[f+'_miscal'] for f in BIAS_LIST})
fm_mis  = pd.Series({f: ft_means[f+'_miscal'] for f in BIAS_LIST})
fs_mis  = pd.Series({f: ft_stds[f+'_miscal'] for f in BIAS_LIST})

order = BIAS_LIST
x = np.arange(len(order))

# 1) Skew: Human vs Base
fig, ax = plt.subplots(figsize=(8,5))
w = 0.35
ax.bar(x-w/2, [human_skew[f] for f in order], w, label='Human', color=cmap(0))
ax.bar(x+w/2, bm_skew[order], w, yerr=bs_skew[order], capsize=5, label='Base', color=cmap(1))
for i,f in enumerate(order): ax.hlines(human_skew[f], i, i+w, colors='green', linestyles='dashed', linewidth=2)
ax.set_xticks(x); ax.set_xticklabels(order, rotation=45, ha='right')
ax.set_ylabel('Skew (% favoring perturbed)'); ax.set_ylim(0,100)
ax.set_title('Skew: Human vs Base'); ax.legend(bbox_to_anchor=(1,1))
plt.tight_layout(); plt.savefig('skew_base_joint.pdf', dpi=300); plt.show()

# 2) Miscalibration: Base only
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(x, bm_mis[order], w, yerr=bs_mis[order], capsize=5, color=cmap(1), label='Base')
ax.set_xticks(x); ax.set_xticklabels(order, rotation=45, ha='right')
ax.set_ylabel('Miscalibration (% disagreement)'); ax.set_ylim(0,100)
ax.set_title('Miscalibration: Base'); ax.legend(bbox_to_anchor=(1,1))
plt.tight_layout(); plt.savefig('miscalibration_base_joint.pdf', dpi=300); plt.show()

# 3) Skew: Human vs Base vs Fine-tuned
fig, ax = plt.subplots(figsize=(8,5))
w2 = 0.25
ax.bar(x-w2, [human_skew[f] for f in order], w2, label='Human', color=cmap(0))
ax.bar(x, bm_skew[order], w2, yerr=bs_skew[order], capsize=5, label='Base', color=cmap(1))
ax.bar(x+w2, fm_skew[order], w2, yerr=fs_skew[order], capsize=5, label='Fine-tuned', color=cmap(2))
for i,f in enumerate(order): ax.hlines(human_skew[f], i - w2 * 0.5, i+ w2 * 1.5, colors='green', linestyles='dashed', linewidth=2)
ax.set_xticks(x); ax.set_xticklabels(order, rotation=45, ha='right')
ax.set_ylabel('Skew (% favoring perturbed)'); ax.set_ylim(0,100)
ax.set_title('Skew: Human vs Base vs Fine-tuned'); ax.legend(bbox_to_anchor=(1,1))
plt.tight_layout(); plt.savefig('skew_ft_joint.pdf', dpi=300); plt.show()

# 4) Miscalibration: Base vs Fine-tuned
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(x-w/2, bm_mis[order], w, yerr=bs_mis[order], capsize=5, label='Base', color=cmap(1))
ax.bar(x+w/2, fm_mis[order], w, yerr=fs_mis[order], capsize=5, label='Fine-tuned', color=cmap(2))
ax.set_xticks(x); ax.set_xticklabels(order, rotation=45, ha='right')
ax.set_ylabel('Miscalibration (% disagreement)'); ax.set_ylim(0,100)
ax.set_title('Miscalibration: Base vs Fine-tuned'); ax.legend(bbox_to_anchor=(1,1))
plt.tight_layout(); plt.savefig('miscalibration_ft_joint.pdf', dpi=300); plt.show()

print('Skew means human:', human_skew)
print('Base skew:', bm_skew.to_dict())
print('FT skew:', fm_skew.to_dict())
print('Base miscal:', bm_mis.to_dict())
print('FT miscal:', fm_mis.to_dict())
```

```python
# === Aggregate differences and print summaries ===
# Miscalibration reduction from base to fine-tuned
miscal_diffs = bm_mis - fm_mis
print("Average miscalibration reduction from base to fine-tuned: {:.4f}%".format(miscal_diffs.mean()))

# Overall means
print("Overall average base miscalibration: {:.4f}%".format(bm_mis.mean()))
print("Overall average fine-tuned miscalibration: {:.4f}%".format(fm_mis.mean()))
print("Overall average base skew: {:.4f}%".format(bm_skew.mean()))
print("Overall average fine-tuned skew: {:.4f}%".format(fm_skew.mean()))

# Differences to human skew
base_diffs = [abs(human_skew[b] - bm_skew[b]) for b in order]
ft_diffs   = [abs(human_skew[b] - fm_skew[b]) for b in order]
avg_base_diff = np.mean(base_diffs)
avg_ft_diff   = np.mean(ft_diffs)
print("Average absolute difference from human skew (base): {:.4f}%".format(avg_base_diff))
print("Average absolute difference from human skew (fine-tuned): {:.4f}%".format(avg_ft_diff))
```
